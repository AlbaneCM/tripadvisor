```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
from time import sleep
from random import randint
```

```python
div_hotels = []
listing_title = []
is_sponsored = []
listing_id = []
listing_url = []
number_reviews = []
listing_rank = []
all_rank = []
common_icons = []
is_special_offer = []

from_rate = []
deal_rate = []
discount_deal = []


other_deals = []

currency = []
```

```python
pages = np.arange(0, 811, 30)

for page in pages:
    page = requests.get('https://www.tripadvisor.com/Hotels-g60763-oa' + str(page) + '-New_York_City_New_York-Hotels.html')
    soup = BeautifulSoup(page.text, 'html.parser')
    
    hotel_divs = soup.find_all('div', class_=['prw_rup prw_meta_hsx_responsive_listing ui_section listItem'])
    
    sleep(randint(1,3))
    
    for result in hotel_divs:
    #     making sure number of property/divs is correctly defined 
        div_hotels.append("property")

    #     getting property title/hotel name
        title = result.find('a', attrs={'class':'property_title'}).text
        listing_title.append(title)

    # getting if hotel is sponsored or not 
        sponsored_check = result.find('span', attrs={'class': 'ui_merchandising_pill sponsored_v2'})
        if sponsored_check == None:
            sponsored_check = False
        else: 
            sponsored_check = True
        is_sponsored.append(sponsored_check)

    # getting hotel id
        hotel_id = result.find('a', attrs={'class': 'property_title prominent'})['id']
        hotel_id = int(hotel_id[9:])
        listing_id.append(hotel_id)

    # getting hotel url
        hotel_url = result.find('a', attrs={'class': 'property_title prominent'})['href']
        hotel_url = 'https://www.tripadvisor.com' + hotel_url
        listing_url.append(hotel_url)

    # getting number of reviews
        review_count = result.find('a', attrs={'class': 'review_count'}).text
        review_count = int(review_count.split(" ")[0].replace(',',''))
        number_reviews.append(review_count)

    # getting listing rank - both hotel and total 
        rank_num = result.find('div', attrs={'class': 'popindex'})
        if rank_num == None:
            rank_num = False
            rank_total = False
        else:
    #         hotel
            rank_num = int(result.find('div', attrs={'class': 'popindex'}).text.split(" ")[0][1:])
    #         total
            rank_total = result.find('div', attrs={'class': 'popindex'}).text.split(" ")[4]
        listing_rank.append(rank_num)
        all_rank.append(rank_total)

    # getting common icons
        icon_list = result.find('ul', attrs={'class': 'icons_list easyClear vertical'}).text
    #     print(icon_list)
        common_icons.append(icon_list)

    # does hotel offer special offer
        special_link = result.find('span', attrs={'class': 'name'})
        if special_link == None:
            special_link = False
        else:
            special_link = True
        is_special_offer.append(special_link)

    # hotel deals: from rate
        rates = result.find('div', attrs={'class': 'price-wrap'})
    #     print(rates)
        if rates != None:
            crossed_rates = rates.text.split("€")
            if len(crossed_rates) == 3:
                crossed_rates = crossed_rates[1]
                from_rate.append(crossed_rates)
            else:
                crossed_rates = 0
                from_rate.append(crossed_rates)
        else:
            crossed_rates = False
            from_rate.append(crossed_rates)
    #     print(crossed_rates)     
```

```python
# verifying that all columns pull same number of rows each time
print(len(div_hotels))
print(len(listing_title))
print(len(is_sponsored))
print(len(listing_id))
print(len(listing_url))
print(len(number_reviews))
print(len(listing_rank))
print(len(all_rank))
print(len(common_icons))
print(len(is_special_offer))
print(len(from_rate))
```
